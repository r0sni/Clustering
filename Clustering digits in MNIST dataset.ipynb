```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.datasets import fetch_openml
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA
```

```python
# Fetch the MNIST dataset
data = fetch_openml('mnist_784', version=1, return_X_y=True)
```

```python
# Split the data into features (X) and labels (y)
X, y = data
print(X.shape)
print(y.shape)
```

```python
#Display the first image
#Reshape the data from 1D array to 2D array (28x28 pixels)
image = X.iloc[0].values.reshape(28, 28)

# Plot the image
plt.imshow(image, cmap='gray')
plt.title(f'Label: {y[0]}')
plt.axis('off')  # Hide the axis
plt.show()
```

```python
#Sample the data for quick visualization
sample_size = 2500
X_sample = X.sample(n=sample_size, random_state=42)
y = y.astype(int)
y_sample = y[X_sample.index]
```

## Clustering using t-SNE

```python
#Apply t_SNE
tsne = TSNE(n_components=2, perplexity=30, learning_rate=10, max_iter=10000, random_state=42)
X_tsne = tsne.fit_transform(X_sample)

#Plot the t-SNE result
plt.figure(figsize=(10,7))
sns.scatterplot(x=X_tsne[:,0], y=X_tsne[:,1], hue=y_sample, palette='deep',legend='full', s=30)
plt.title('t-SNE Visualization of MNIST Dataset')
plt.xlabel('t-SNE Feature 1')
plt.ylabel('t-SNE Feature 2')
plt.legend(title='Digits')
plt.show()
```
